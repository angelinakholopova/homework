## Задание 2
- Постройте график
- Сделайте выводы

### 2.1 Изучите распределение домов от наличия вида на набережную

```python
df.loc[df['waterfront'] == 0, 'waterfront'] = 'no'
df.loc[df['waterfront'] == 1, 'waterfront'] = 'yes'
waterfront = df['waterfront'].value_counts()
names = waterfront.index
values = waterfront.values
plt.pie(values, labels = names, autopct = '%.1f%%');
```

### 2.2 Изучите распределение этажей домов

```python
weight = df['floors'].value_counts()
names = weight.index
values = weight.values
plt.bar(x = names, height = values);
plt.xlabel('floor')
plt.ylabel('quantity')
plt.title('floor bar');
```

### 2.3 Изучите распределение состояния домов


```python
weight = df['condition'].value_counts()
names = weight.index
values = weight.values
plt.bar(x = names, height = values);
plt.xlabel('condition')
plt.ylabel('quantity')
plt.title('bar condition');
```
